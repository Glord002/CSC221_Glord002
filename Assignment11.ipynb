```python
#9.1
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        """
        Initialize the Restaurant class with a restaurant name and cuisine type.

        Parameters:
        restaurant_name (str): The name of the restaurant.
        cuisine_type (str): The type of cuisine served by the restaurant.
        """
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        """Print the name and cuisine type of the restaurant."""
        print(f"Restaurant Name: {self.restaurant_name}")
        print(f"Cuisine Type: {self.cuisine_type}")

    def open_restaurant(self):
        """Print a message indicating that the restaurant is open."""
        print(f"{self.restaurant_name} is now open!")


restaurant = Restaurant("La Trattoria", "Italian")

print("Restaurant Name:", restaurant.restaurant_name)
print("Cuisine Type:", restaurant.cuisine_type)

restaurant.describe_restaurant()
restaurant.open_restaurant()

#9.2
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        """
        Initialize the Restaurant class with a restaurant name and cuisine type.

        Parameters:
        restaurant_name (str): The name of the restaurant.
        cuisine_type (str): The type of cuisine served by the restaurant.
        """
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        """Print the name and cuisine type of the restaurant."""
        print(f"Restaurant Name: {self.restaurant_name}")
        print(f"Cuisine Type: {self.cuisine_type}")


restaurant1 = Restaurant("La Trattoria", "Italian")
restaurant2 = Restaurant("The Sushi Bar", "Japanese")
restaurant3 = Restaurant("Spice Route", "Indian")

restaurant1.describe_restaurant()
restaurant2.describe_restaurant()
restaurant3.describe_restaurant()

#9.3
class User:
    def __init__(self, first_name, last_name, age, email, country):
        """
        Initialize the User class with attributes for a user profile.

        Parameters:
        first_name (str): The first name of the user.
        last_name (str): The last name of the user.
        age (int): The age of the user.
        email (str): The email address of the user.
        country (str): The country of residence of the user.
        """
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email
        self.country = country

    def describe_user(self):
        """Print a summary of the user's information."""
        print(f"User Information:")
        print(f"First Name: {self.first_name}")
        print(f"Last Name: {self.last_name}")
        print(f"Age: {self.age}")
        print(f"Email: {self.email}")
        print(f"Country: {self.country}")

    def greet_user(self):
        """Print a personalized greeting to the user."""
        print(f"Hello, {self.first_name}! Welcome back.")


user1 = User("George", "Lordkipanidze", 20, "georgelortki@gmail.com", "Delisi")
user2 = User("Nick", "Javaxishvili", 21, "Nika.javakha93@gmail.com", "Delisi")
user3 = User("David", "Shvelidze", 22, "Datoshveli@gmail.com", "Delisi")

user1.describe_user()
user1.greet_user()
print()
user2.describe_user()
user2.greet_user()
print()
user3.describe_user()
user3.greet_user()

#9.6
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        """
        Initialize the Restaurant class with a restaurant name and cuisine type.

        Parameters:
        restaurant_name (str): The name of the restaurant.
        cuisine_type (str): The type of cuisine served by the restaurant.
        """
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        """Print the name and cuisine type of the restaurant."""
        print(f"Restaurant Name: {self.restaurant_name}")
        print(f"Cuisine Type: {self.cuisine_type}")

    def open_restaurant(self):
        """Print a message indicating that the restaurant is open."""
        print(f"{self.restaurant_name} is now open!")


class IceCreamStand(Restaurant):
    def __init__(self, restaurant_name, cuisine_type, flavors):
        """
        Initialize the IceCreamStand class.

        Parameters:
        restaurant_name (str): The name of the ice cream stand.
        cuisine_type (str): The type of cuisine served by the ice cream stand (usually 'Ice Cream').
        flavors (list): List of ice cream flavors offered by the stand.
        """
        super().__init__(restaurant_name, cuisine_type)
        self.flavors = flavors

    def display_flavors(self):
        """Print the list of ice cream flavors offered by the stand."""
        print(f"Ice Cream Flavors:")
        for flavor in self.flavors:
            print(flavor)


ice_cream_stand = IceCreamStand("Scoops & Smiles", "Ice Cream", ["Vanilla", "Chocolate", "Strawberry", "Mint"])

ice_cream_stand.display_flavors()

#9.7
class User:
    def __init__(self, first_name, last_name, age, email, country):
        """
        Initialize the User class with attributes for a user profile.

        Parameters:
        first_name (str): The first name of the user.
        last_name (str): The last name of the user.
        age (int): The age of the user.
        email (str): The email address of the user.
        country (str): The country of residence of the user.
        """
        self.first_name = first_name
        self.last_name = last_name
        self.age = age
        self.email = email
        self.country = country

    def describe_user(self):
        """Print a summary of the user's information."""
        print(f"User Information:")
        print(f"First Name: {self.first_name}")
        print(f"Last Name: {self.last_name}")
        print(f"Age: {self.age}")
        print(f"Email: {self.email}")
        print(f"Country: {self.country}")

    def greet_user(self):
        """Print a personalized greeting to the user."""
        print(f"Hello, {self.first_name}! Welcome back.")


class Admin(User):
    def __init__(self, first_name, last_name, age, email, country, privileges):
        """
        Initialize the Admin class.

        Parameters:
        first_name (str): The first name of the admin.
        last_name (str): The last name of the admin.
        age (int): The age of the admin.
        email (str): The email address of the admin.
        country (str): The country of residence of the admin.
        privileges (list): List of privileges of the admin.
        """
        super().__init__(first_name, last_name, age, email, country)
        self.privileges = privileges

    def show_privileges(self):
        """Print the list of privileges of the admin."""
        print("Admin Privileges:")
        for privilege in self.privileges:
            print(privilege)


admin = Admin("George", "Lordkipanidze", 20, "georgelortki@gmail.com", "Delisi", ["can add post", "can delete post", "can ban user"])

admin.show_privileges()
```
