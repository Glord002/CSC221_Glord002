```python
names = ['George', 'Nick', 'David', 'Tamara']

for name in names:
    print(name)

for name in names:
    print(f"Hello, {name}! How are you today?")

transportation = ['car', 'motorcycle', 'bicycle', 'train']

for item in transportation:
    print(f"I would like to own a {item}.")
guests = ['Albert Einstein', 'Leonardo da Vinci', 'Marie Curie']

for guest in guests:
    print(f"Dear {guest},\n\tYou are cordially invited to dinner. Please join us for a delightful evening.\n")

guest_unavailable = 'Albert Einstein'
print(f"{guest_unavailable} can't make it to the dinner.")

guests[guests.index(guest_unavailable)] = 'Nikola Tesla'

for guest in guests:
    print(f"Dear {guest},\n\tYou are cordially invited to dinner. Please join us for a delightful evening.\n")

print("Good news! We found a bigger dinner table.")

guests.insert(0, 'Isaac Newton')
guests.insert(len(guests) // 2, 'Galileo Galilei')
guests.append('Stephen Hawking')

for guest in guests:
    print(f"Dear {guest},\n\tYou are cordially invited to dinner. Please join us for a delightful evening.\n")

print("Unfortunately, the new dinner table won't arrive in time for the dinner. We can only invite two people.")

while len(guests) > 2:
    removed_guest = guests.pop()
    print(f"Sorry {removed_guest}, but we won't be able to invite you to dinner.")

for remaining_guest in guests:
    print(f"Dear {remaining_guest},\n\tYou are still invited to dinner. Please join us for a delightful evening.\n")

del guests[:]
print("The guest list is now empty:", guests)

places_to_visit = ['Japan', 'Italy', 'Australia', 'Egypt', 'Canada']

print("Original order:")
print(places_to_visit)

print("\nSorted in alphabetical order:")
print(sorted(places_to_visit))

print("\nOriginal order (still):")
print(places_to_visit)

print("\nSorted in reverse alphabetical order:")
print(sorted(places_to_visit, reverse=True))

print("\nOriginal order (still):")
print(places_to_visit)

places_to_visit.reverse()
print("\nReversed order:")
print(places_to_visit)

places_to_visit.reverse()
print("\nBack to original order:")
print(places_to_visit)

places_to_visit.sort()
print("\nSorted in alphabetical order:")
print(places_to_visit)

places_to_visit.sort(reverse=True)
print("\nSorted in reverse alphabetical order:")
print(places_to_visit)

print("\nNumber of people invited to dinner:", len(places_to_visit))

print("\nUsing each function introduced in Chapter 3:")
my_list = ['apple', 'banana', 'cherry']

my_list.append('date')
print("After appending:", my_list)

my_list.insert(2, 'grape')
print("After inserting:", my_list)

popped_item = my_list.pop()
print("Popped item:", popped_item)
print("After popping:", my_list)

my_list.remove('banana')
print("After removing:", my_list)

my_list.sort()
print("After sorting:", my_list)

my_list.reverse()
print("After reversing:", my_list)

print("Index of 'apple':", my_list.index('apple'))

print("Count of 'cherry':", my_list.count('cherry'))

my_list.clear()
print("After clearing:", my_list)

my_list = [1, 2, 3, 4, 5]

try:
    print(my_list[10])
except IndexError:
    print("Index error occurred!")
```
